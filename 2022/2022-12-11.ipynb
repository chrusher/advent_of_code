```python
test_input = '''Monkey 0:
  Starting items: 79, 98
  Operation: new = old * 19
  Test: divisible by 23
    If true: throw to monkey 2
    If false: throw to monkey 3

Monkey 1:
  Starting items: 54, 65, 75, 74
  Operation: new = old + 6
  Test: divisible by 19
    If true: throw to monkey 2
    If false: throw to monkey 0

Monkey 2:
  Starting items: 79, 60, 97
  Operation: new = old * old
  Test: divisible by 13
    If true: throw to monkey 1
    If false: throw to monkey 3

Monkey 3:
  Starting items: 74
  Operation: new = old + 3
  Test: divisible by 17
    If true: throw to monkey 0
    If false: throw to monkey 1'''
```

```python
#part 1
monkeys = {}
monkey = ''

for line in test_input.split('\n'):
    if 'Monkey' in line:
        monkey = line.split()[-1].strip(':')
        monkeys[monkey] = {}
        
    elif 'Starting' in line:
        monkeys[monkey]['items'] = [int(item.strip()) for item in line.split(':')[-1].split(',')]

    elif 'Operation' in line:
        monkeys[monkey]['operation'] = line.split()[-2:]
    
    elif 'Test' in line:
        monkeys[monkey]['test'] = int(line.split()[-1])
    
    elif 'If true' in line:
        monkeys[monkey]['true'] = line.split()[-1]
    
    elif 'If false' in line:
        monkeys[monkey]['false'] = line.split()[-1]
        
        
print(monkeys)

def operation(x, args):
    
    if args[0] == '+':
        return x + int(args[1])
    elif args[0] == '*':
        if args[1] == 'old':
            return x * x
        else:
            return x * int(args[1])

counts = {monkey:0 for monkey in monkeys}        
        
for i in range(20):
    for monkey in monkeys:
    #     print(monkey)

        for item in monkeys[monkey]['items']:
            counts[monkey] += 1
            level = operation(item, monkeys[monkey]['operation']) // 3
            if not (level % monkeys[monkey]['test']):
                monkeys[monkeys[monkey]['true']]['items'].append(level)
    #             print('throwing', level, 'to', monkeys[monkey]['true'])
            else:
                monkeys[monkeys[monkey]['false']]['items'].append(level)
    #             print('throwing', level, 'to', monkeys[monkey]['false'])       

        monkeys[monkey]['items'] = []        
    #     print()

    print('Round', i + 1)
    for monkey in monkeys:
        print(monkey, ':', monkeys[monkey]['items'])
        
sorted_counts = sorted(counts.values())
print(sorted_counts)
print(sorted_counts[-1] * sorted_counts[-2])
```
