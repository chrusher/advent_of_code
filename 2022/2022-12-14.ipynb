```python
test_input = '''498,4 -> 498,6 -> 496,6
503,4 -> 502,4 -> 502,9 -> 494,9'''
```

```python
def min_max(grid):
    min_x = 10000
    max_x = 0
    min_y = 10000
    max_y = 0

    for point in grid:
        if point[0] < min_x:
            min_x = point[0]
        if point[0] > max_x:
            max_x = point[0]
        if point[1] < min_y:
            min_y = point[1]
        if point[1] > max_y:
            max_y = point[1]
            
    return min_x, max_x, min_y, max_y

def print_grid(grid):

    min_x, max_x, min_y, max_y = min_max(grid)
    output = ''
    for i in range(min_y, max_y + 1):
        row = ''
        for j in range(min_x, max_x + 1):
            if (j, i) in grid:
                row = row + grid[(j, i)]
            else:
                row = row + '.'
        output = output + row + '\n'

    print(output)   
    

```

```python
#part 1
start = (500, 0)
grid = {start:'+'}

lines = test_input.split('\n')
lines = open('2022-12-14_input').readlines()
for line in lines:
    line = line.strip()
    
    points = [[int(coord) for coord in point.split(',')] for point in line.split(' -> ')]

    grid[tuple(points[0])] = '#'
    
    for i in range(1, len(points)):
        if points[i - 1][0] < points[i][0]:
            for j in range(points[i - 1][0], points[i][0] + 1):
                grid[(j, points[i][1])] = '#'
        if points[i - 1][0] > points[i][0]:
            for j in range(points[i][0], points[i - 1][0]):
                grid[(j, points[i][1])] = '#'
        if points[i - 1][1] < points[i][1]:
            for j in range(points[i - 1][1], points[i][1] + 1):
                grid[(points[i][0]), j] = '#'
        if points[i - 1][1] > points[i][1]:
            for j in range(points[i][1], points[i - 1][1]):
                grid[(points[i][0]), j] = '#'            

min_x, max_x, min_y, max_y = min_max(grid)
print_grid(grid)

count = 0
flowing = False  
while not flowing:
    sand_coord = list(start)
    falling = True
    while falling:
        if sand_coord[1] > max_y:
            flowing = True
            break

        if (sand_coord[0], sand_coord[1] + 1) not in grid:
            sand_coord[1] += 1
        elif (sand_coord[0] - 1, sand_coord[1] + 1) not in grid:
            sand_coord[0] += -1
            sand_coord[1] += 1
        elif (sand_coord[0] + 1, sand_coord[1] + 1) not in grid:
            sand_coord[0] += 1
            sand_coord[1] += 1
        else:
            grid[tuple(sand_coord)] = 'o'
            falling = False
            count += 1
            
print(count)
print_grid(grid)


```
