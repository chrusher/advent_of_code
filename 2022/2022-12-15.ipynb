```python
import re
import sys
```

```python
test_input = '''Sensor at x=2, y=18: closest beacon is at x=-2, y=15
Sensor at x=9, y=16: closest beacon is at x=10, y=16
Sensor at x=13, y=2: closest beacon is at x=15, y=3
Sensor at x=12, y=14: closest beacon is at x=10, y=16
Sensor at x=10, y=20: closest beacon is at x=10, y=16
Sensor at x=14, y=17: closest beacon is at x=10, y=16
Sensor at x=8, y=7: closest beacon is at x=2, y=10
Sensor at x=2, y=0: closest beacon is at x=2, y=10
Sensor at x=0, y=11: closest beacon is at x=2, y=10
Sensor at x=20, y=14: closest beacon is at x=25, y=17
Sensor at x=17, y=20: closest beacon is at x=21, y=22
Sensor at x=16, y=7: closest beacon is at x=15, y=3
Sensor at x=14, y=3: closest beacon is at x=15, y=3
Sensor at x=20, y=1: closest beacon is at x=15, y=3'''
```

```python
full_input = open('2022-12-15_input').read()
print(full_input)
```

```python
def metric(ax, ay, bx, by):
    return abs(ax - bx) + abs(ay - by)
```

```python
sensors = {}
beacons = set()

for line in test_input.split('\n'):
    sensor, beacon = re.findall('x=(-?\\d+), y=(-?\\d+)', line)
    sx, sy = sensor
    sx = int(sx)
    sy = int(sy)
    bx, by = beacon
    bx = int(bx)
    by = int(by)
    
    sensors[(sx, sy)] = metric(sx, sy, bx, by)
    beacons.add((bx, by))
    
        
output = ''
for i in range(0, 23):        
    for j in range(-2, 26):       
        if (j, i) in sensors:
            output += 'S'
        elif (j, i) in beacons:
            output += 'B'                  
        else:
            output += '.'
    output += '\n'
print(output)

print(sensors)

output = ''
for i in range(9, 12):  
    row_count = 0
    for j in range(-4, 27):
        if (j, i) in sensors:
            s = 'S'
        elif (j, i) in beacons:
            s = 'B'
        else:
            s = '.'
            for sensor in sensors:
                if metric(sensor[0], sensor[1], j, i) <= sensors[sensor]:
                    s = '#'
        if s not in  ['.', 'B']:
            row_count += 1
        output += s
    print(i, row_count)
    output += '\n'
print(output)

i = 10
limits = []
for sensor in sensors:
    distance = metric(sensor[0], sensor[1], sensor[0], i)
    if distance <= sensors[sensor]:
        limits.append([sensor[0] - (sensors[sensor] - distance), sensor[0] + (sensors[sensor] - distance)])

beacon_count = 0
for beacon in beacons:
    if beacon[1] == i:
        beacon_count += 1
        
limits.sort(key=lambda x:x[0])
new_limits = [limits[0]]
for limit in limits[1:]:
    if limit[0] <= new_limits[-1][1]:
        new_limits[-1][1] = max(new_limits[-1][1], limit[1])
    else:
        new_limits.append(limit)
        
count = -beacon_count
for limit in new_limits:
    count += limit[1] - limit[0] + 1
        
print(count)
print()


output = ''
for i in range(0, 21):  
    for j in range(0, 21):
        if (j, i) in sensors:
            s = 'S'
        elif (j, i) in beacons:
            s = 'B'
        else:
            s = '.'
            for sensor in sensors:
                if metric(sensor[0], sensor[1], j, i) <= sensors[sensor]:
                    s = '#'
        if s == '.':
            print(j, i)
        output += s
    output += '\n'
print(output)

for i in range(0, 20):
    limits = []
    for sensor in sensors:
        distance = metric(sensor[0], sensor[1], sensor[0], i)
        if distance <= sensors[sensor]:
            limits.append([sensor[0] - (sensors[sensor] - distance), sensor[0] + (sensors[sensor] - distance)])
            
    limits.sort(key=lambda x:x[0])
    new_limits = [limits[0]]
    for limit in limits[1:]:
        if limit[0] <= new_limits[-1][1] + 1:
            new_limits[-1][1] = max(new_limits[-1][1], limit[1])
        else:
            new_limits.append(limit)
            
    print(new_limits)
    
    limits = []
    for limit in new_limits:
        if limit[1] > 0 and limit[0] <= 20:
            limits.append([max(0, limit[0]), min(limit[1], 20)])
            
    if len(limits) > 1:
        print(limits[0][1] + 1, i)
        print((limits[0][1] + 1) * 4000000 + i)
```

```python
sensors = {}
beacons = set()

max_distance = 0
min_x = sys.maxsize
max_x = -sys.maxsize - 1
for line in full_input.split('\n'):
    if not line:
        continue
    sensor, beacon = re.findall('x=(-?\\d+), y=(-?\\d+)', line)
    sx, sy = sensor
    sx = int(sx)
    sy = int(sy)
    if sx > max_x:
        max_x = sx
    if sx < min_x:
        min_x = sx
    
    bx, by = beacon
    bx = int(bx)
    by = int(by)
    
    distance = metric(sx, sy, bx, by)
    sensors[(sx, sy)] = distance
    beacons.add((bx, by))
    if distance > max_distance:
        max_distance = distance
           

i = 2000000
limits = []
for sensor in sensors:
    distance = metric(sensor[0], sensor[1], sensor[0], i)
    if distance <= sensors[sensor]:
        limits.append([sensor[0] - (sensors[sensor] - distance), sensor[0] + (sensors[sensor] - distance)])

beacon_count = 0
for beacon in beacons:
    if beacon[1] == i:
        beacon_count += 1
        
limits.sort(key=lambda x:x[0])
new_limits = [limits[0]]
for limit in limits[1:]:
    if limit[0] <= new_limits[-1][1] + 1:
        new_limits[-1][1] = max(new_limits[-1][1], limit[1])
    else:
        new_limits.append(limit)
        
count = -beacon_count
for limit in new_limits:
    count += limit[1] - limit[0] + 1
        
print(count)
```
