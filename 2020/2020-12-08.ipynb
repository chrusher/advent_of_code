```python
accumulator = 0
stack = 0
excuted = set()

def run_program(program):
    accumulator = 0
    stack = 0
    excuted = set()    

    while True:    
#         print(stack, accumulator)
        if stack in excuted:
            return False, accumulator
        excuted.add(stack)

        op, arg = program[stack]

        if op == 'acc':
            accumulator += arg
            stack += 1
        elif op == 'jmp':
            stack += arg
        else:
            stack += 1
            
        if stack >= len(program):
            return True, accumulator
        
print(run_program(program))
            
    
for i in range(1, len(program)):
    new_program = program.copy()
    
    if 'jmp' == new_program[i][0]:
        new_program[i] = ('nop', new_program[i][1])
    elif 'nop' == new_program[i][0]:
        new_program[i] = ('jmp', new_program[i][1])
    
    result, count = run_program(new_program)
    if result:
        print(count)
        break
         
        
```
