```python
# print(addapters)
def find_addapters(start, end):   
    finished = []
    def _find_addapters(start, end):
        if start == end:
            finished.append(True)
            return

        possibles = [addapter for addapter in addapters if (addapter - start <= 3) and (addapter > start)]
        for possible in possibles:
            _find_addapters(possible, end)
            
    _find_addapters(start, end)        
    return len(finished)

print(find_addapters(0, 1))
print(find_addapters(4, 7))
print(find_addapters(10, 12))
print(find_addapters(15, 16))
```
