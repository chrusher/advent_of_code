```python
def compute(string):
#     print(string, end=' ')
    while True:
        matches = re.findall('(\d+ [+*] \d+)', string)
#         print(matches)
        if not matches:
            break
        result = str(eval(matches[0]))
        string = string.replace(matches[0], result, 1)    
#     print(string)
    return string


def compute_brackets(string):
    while True:
        matches = re.findall('\([^(^)]+\)', string)
        if not matches:
            break
        for match in matches:
            string = string.replace(match, compute(match[1:-1]), 1)
    return compute(string)


def compute_addition(string):
    while True:
        matches = re.findall('(\d+ \+ \d+)', string)
        if not matches:
            break
        result = str(eval(matches[0]))
        string = string.replace(matches[0], result, 1)
        
    while True:
        matches = re.findall('(\d+ \* \d+)', string)
        if not matches:
            break
        result = str(eval(matches[0]))
        string = string.replace(matches[0], result, 1)  
        
    return string


def compute_brackets_addition(string):
    while True:
        matches = re.findall('\([^(^)]+\)', string)
        if not matches:
            break
        for match in matches:
            string = string.replace(match, compute_addition(match[1:-1]), 1)
    return compute_addition(string)
```
