```python
def mapping(values, rules):
    
    results = []
    for value in values:
        for rule in rules:
            destination, source, length = rule
            if source <= value < source + length:
                results.append(value + destination - source)
                break
        else:
            results.append(value)
    return results

def normalise_rules(rules):
    
    rules = sorted(rules, key=lambda rule: rule[1])
    
    for i in range(len(rules) - 1):
        start = rules[i][1] + rules[i][2]
        end = rules[i + 1][1]
        
        if start != end:
            rules = rules[:i + 1] + [[start, start, end - start]] + rules[i + 1:]
    
    if rules[0][1] != 0:
        rules = [[0, 0, rules[0][1]]] + rules
    if rules[-1][2] != float('inf'):    
        rules = rules + [[rules[-1][1] + rules[-1][2], rules[-1][1] + rules[-1][2], float('inf')]]

    return rules

def combine_rules(first_rules, second_rules):
    new_rules = []
    for rule_1 in first_rules:
        dest_1, source_1, len_1 = rule_1
        end_1 = dest_1 + len_1

        for rule_2 in second_rules:
            dest_2, source_2, len_2 = rule_2
            end_2 = source_2 + len_2
            
            overlap = end_1 > source_2 and dest_1 < end_2
            if overlap:
                new_source = max(dest_1, source_2)            
                new_len = min(end_1, end_2) - new_source
                new_dest = dest_2 - source_2 + new_source
                new_source = new_source + source_1 - dest_1
                if new_len:
                    new_rules.append([new_dest, new_source, new_len])

    new_rules.sort(key=lambda rule: rule[1])   
    return(new_rules)
```

```python
# part 2

maps = []

for line in lines[2:]:
    if 'map' in line:
        name = line.split()[0]
        maps.append([])
    elif line:
        maps[-1].append([int(number) for number in line.split()])
        
new_rules = normalise_rules(maps[0])

for i in range(1, len(maps)):   
    new_rules = combine_rules(new_rules, normalise_rules(maps[i]))
      
    
seeds = [int(seed) for seed in lines[0].split(':')[-1].split()]

print(min(mapping(seeds, new_rules)))
print()

minima = float('inf')
for i in range(0, len(seeds), 2):
    
    start = seeds[i]
    end = start + seeds[i + 1]
    
    for rule in new_rules:
        dest, source, length = rule
        
        overlap = end > source and start < source + length
        if overlap:
            if source >= start:
                new_dest = dest
            else:
                new_dest = dest - source + start
            if new_dest < minima:

                minima = new_dest

    
print(minima)
```
