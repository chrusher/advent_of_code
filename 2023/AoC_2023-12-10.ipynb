```python
lines = [line.strip() for line in open('AoC_2023-12-10_input').readlines()]
```

```python
start = None
for i in range(len(lines)):
    for j in range(len(lines[0])):
        if lines[i][j] == 'S':
            start = (i, j)
            
print('Start', start)
print()
graph = {}
adjacent = []

if start[0] - 1 >= 0 and lines[start[0] - 1][start[1]] in ['|', '7', 'F']:
    adjacent.append('N')
if start[0] + 1 < len(lines) and lines[start[0] + 1][start[1]] in ['|', 'L', 'J']:
    adjacent.append('S')
if start[1] - 1 >= 0 and lines[start[0]][start[1] - 1] in ['-', 'L', 'F']:
    adjacent.append('W')
if start[1] + 1 < len(lines[0]) and lines[start[0]][start[1] + 1] in ['-', 'J', '7']:
    adjacent.append('E')    
    

# pipes = {'|':('N', 'S'), '-':('E', 'W'), 'L':('N', 'E'), 'J':('N', 'W'), '7':('S', 'W'), 'F':('S', 'E')}
directions = {'N':(-1, 0), 'E':(0, 1), 'S':(1, 0), 'W':(0, -1)}

pipes = {'|':{'S':'S', 'N':'N'},
         '-':{'E':'E', 'W':'W'},
         'L':{'S':'E', 'W':'N'},
         'J':{'S':'W', 'E':'N'},
         '7':{'N':'W', 'E':'S'}, 
         'F':{'N':'E', 'W':'S'}}

# print(adjacent)
direction = adjacent[0]
point = (start[0] + directions[direction][0], start[1] + directions[direction][1])
print(start, 'S', direction)
steps = 0
loop= [start, point]

while True:

    symbol = lines[point[0]][point[1]]    
    if symbol == 'S':
        break
    direction = pipes[symbol][direction]
#     print(point, symbol, direction)
    point = (point[0] + directions[direction][0], point[1] + directions[direction][1])
    loop.append(point)
    steps += 1

# print(steps)
print('Distance', steps // 2 + 1)

```

```python
loop_string = [['.' for c in range(2 * len(lines[0]) - 1)] for l in range(2 * len(lines) - 1)]

# for i in range(len(lines)):
#     for j in range(len(lines[0])):
#         loop_string[2 * i][2 * j] = 'o'

last_point = loop[0]
for point in loop[1:]:
    loop_string[2 * point[0]][2 * point[1]] = '█'
    loop_string[last_point[0] + point[0]][last_point[1] + point[1]] = '█'
    last_point = point

to_visit = []
for i in range(len(loop_string)):
    if loop_string[i][0] == '.':
        to_visit.append((i, 0))
    if loop_string[i][-1] == '.':
        to_visit.append((i, len(loop_string[0]) - 1))
        
for i in range(1, len(loop_string[0]) - 1):
    if loop_string[0][i] == '.':
        to_visit.append((0, i))
    if loop_string[-1][i] == '.':
        to_visit.append((len(loop_string) - 1, i))
    
    
# to_visit = [(0, 0)]
count = 0
while len(to_visit):
    unvisited = []
    for point in to_visit:
        loop_string[point[0]][point[1]] = 'O'
        if point[0] - 1 > 0 and loop_string[point[0] - 1][point[1]] == '.':
            unvisited.append((point[0] - 1, point[1]))
        if point[0] + 1 < len(loop_string) and loop_string[point[0] + 1][point[1]] == '.':
            unvisited.append((point[0] + 1, point[1]))
        if point[1] - 1 > 0 and loop_string[point[0]][point[1] - 1] == '.':
            unvisited.append((point[0], point[1] - 1))
        if point[1] + 1 < len(loop_string[0]) and loop_string[point[0]][point[1] + 1] == '.':
            unvisited.append((point[0], point[1] + 1))
        count += 1
            
    to_visit = unvisited
#     print('\n'.join([''.join(line) for line in loop_string]))
#     print()
#     print(to_visit)

loop_string = [line[::2] for line in loop_string[::2]]
print(count)       
print('\n'.join([''.join(line) for line in loop_string]))
print()
loop_string = '\n'.join([''.join(line) for line in loop_string])
print(loop_string.count('.'))
```
