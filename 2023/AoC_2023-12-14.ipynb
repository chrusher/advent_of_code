```python
# part 2

lines = full_input.copy()

for i in range(len(lines)):
#     print(lines[i])
    lines[i] = list(lines[i])
# print()

positions = []
for n in range(300):
    
    # north
    for i in range(1, len(lines)):
        for j in range(len(lines[0])):
            k = i - 1
            if lines[i][j] == 'O':
                while k >= 0 and lines[k][j] == '.':
                    lines[k + 1][j] = '.'
                    lines[k][j] = 'O'
                    k += -1

    # west
    for i in range(len(lines)):
        for j in range(1, len(lines)):
            k = j - 1
            if lines[i][j] == 'O':
                while k >= 0 and lines[i][k] == '.':
                    lines[i][k + 1] = '.'
                    lines[i][k] = 'O'
                    k += -1            

    # south
    lines = lines[::-1]
    for i in range(1, len(lines)):
        for j in range(len(lines[0])):
            k = i - 1
            if lines[i][j] == 'O':
                while k >= 0 and lines[k][j] == '.':
                    lines[k + 1][j] = '.'
                    lines[k][j] = 'O'
                    k += -1

    lines = lines[::-1]

    # east
    lines = [line[::-1] for line in lines]
    for i in range(len(lines)):
        for j in range(1, len(lines)):
            k = j - 1
            if lines[i][j] == 'O':
                while k >= 0 and lines[i][k] == '.':
                    lines[i][k + 1] = '.'
                    lines[i][k] = 'O'
                    k += -1            

    
    lines = [line[::-1] for line in lines]
    string = '\n'.join([''.join(line) for line in lines])   
#     print(string)
    total = 0
    for i in range(len(lines)):
        total += lines[i].count('O') * (len(lines) - i)
    print(n, total)
    position = (string, total)
    if position in positions:
        print('repeart', n, positions.index(position))
        start = positions.index(position)
        cycle = n - start
        print()
        break
        
    positions.append(position)
#     print()
    
print(positions[(1000000000 - 1 - start) % cycle + start][1])
```
