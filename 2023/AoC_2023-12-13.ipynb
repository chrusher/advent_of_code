```python
test_input = """#.##..##.
..#.##.#.
##......#
##......#
..#.##.#.
..##..##.
#.#.##.#.

#...##..#
#....#..#
..##..###
#####.##.
#####.##.
..##..###
#....#..#
"""

lines = test_input.split('\n')
```

```python
lines = [line.strip() for line in open('AoC_2023-12-13_input')]
lines += ['']
```

```python
# part 1

total = 0
pattern = []
for line in lines:
    if line:
        pattern.append(line)
        continue
    
    for i in range(len(pattern)):        
        upper = pattern[:i]
        lower = pattern[i:]
        length = min(len(upper), len(lower))
        if length and upper[-length:] == lower[:length][::-1]:
            total += i * 100
            print(i * 100)           
            
    columns = []
    for i in range(len(pattern[0])):
        column = ''.join([row[i] for row in pattern])
        columns.append(column)

    for i in range(len(columns)):
        upper = columns[:i]
        lower = columns[i:]
        length = min(len(upper), len(lower))
        if length and upper[-length:] == lower[:length][::-1]:
            total += i
            print(i)            

    pattern = []        
    print()
        
print(total)
```
