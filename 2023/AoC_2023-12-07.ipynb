```python
# part 1

hands = {'five':[], 'four':[], 'full':[], 'three':[], 'two':[], 'pair':[], 'high':[]}

for line in lines:
    cards, bid = line.split()
    bid = int(bid)
#     print(cards, bid)
    
    counts = {}
    for card in cards:
        if card in counts:
            counts[card] += 1
        else:
            counts[card] = 1
    
    values = list(counts.values())
    if 5 in values:
        hands['five'].append((cards, bid))
    elif 4 in values:
        hands['four'].append((cards, bid))
    elif 3 in values and 2 in values:
        hands['full'].append((cards, bid))
    elif 3 in values:
        hands['three'].append((cards, bid))
    elif values.count(2) == 2:
        hands['two'].append((cards, bid))
    elif 2 in values:
        hands['pair'].append((cards, bid))
    else:
        hands['high'].append((cards, bid))
        
# print(hands)

order = ['A', 'K', 'Q', 'J', 'T', '9', '8', '7', '6', '5', '4', '3', '2']

def calc_order(hand):
    
    total = 0
    for i, card in enumerate(hand[::-1]):
        total += order.index(card) * len(order)**i
    return total

for hand in hands:
    hands[hand].sort(key=lambda cards: calc_order(cards[0]))
    
# print(hands)

total = 0
counter = 0
for hand in hands:
    for cards, bid in hands[hand]:
#         print(cards, bid, len(lines) - counter)
        total += bid * (len(lines) - counter)
        counter += 1
        
#     print()
print(total)        
```
