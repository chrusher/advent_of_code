```python
with open('2021-12-13_input') as f:
    puzzle = f.read().strip()
```

```python
points, folds = puzzle.split('\n\n')
points = points.split()
folds = folds.split('\n')

points = np.array([point.split(',') for point in points], dtype=np.int_)

sheet = np.zeros((points[:,0].max() + 1, points[:,1].max() + 2), dtype=np.int_)

for point in points:
    sheet[point[0], point[1]] = 1
    
print(sheet.T)
print(sheet.shape)



for fold in folds:
    
    print(fold.split('='))
    
    line = int(fold.split('=')[-1])
    
    if 'y=' in fold:
        
        upper = sheet[:,:line]
        lower = sheet[:,line + 1:]
        lower = lower[:,::-1]
        sheet = upper | lower
        print(sheet.T)
        print(sheet.sum())
        
        
    else:
        left = sheet[:line]
        right = sheet[line + 1:]
        right = right[::-1]
        sheet = left | right
        print(sheet.T)
        print(sheet.sum())
        

for line in sheet.T:
    
    string = ''.join([str(n) for n in line])
    
    string = string.replace('0', ' ')
    string = string.replace('1', '█')
    
    print(string)
```
