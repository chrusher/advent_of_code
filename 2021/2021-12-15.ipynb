```python
import datetime
from queue import PriorityQueue

import numpy as np
```

```python
example = """1163751742
1381373672
2136511328
3694931569
7463417111
1319128137
1359912421
3125421639
1293138521
2311944581"""
```

```python
with open('2021-12-15_input') as f:
    puzzle = f.read().strip()
```

```python
#part one
start = datetime.datetime.now()
grid = np.array([list(line) for line in example.split()], dtype=np.int_)

grid = np.array([list(line) for line in puzzle.split()], dtype=np.int_)
print(grid.T)

graph = {}

for i in range(grid.shape[0]):
    for j in range(grid.shape[1]):
        
        
        coord = (i, j, grid[i, j])
        
        if i:
            previous = (i - 1, j)
            if previous in graph:
                graph[previous].append(coord)
            else:
                graph[previous] = [coord]
                
        if i < grid.shape[0] - 1:
            previous = (i + 1, j)
            if previous in graph:
                graph[previous].append(coord)
            else:
                graph[previous] = [coord]
                
        if j:
            previous = (i, j - 1)
            if previous in graph:
                graph[previous].append(coord)
            else:
                graph[previous] = [coord]  
                
        if j < grid.shape[1] - 1:
            previous = (i, j + 1)
            if previous in graph:
                graph[previous].append(coord)
            else:
                graph[previous] = [coord]                
                
# print(graph)

vertices = set(graph.keys())
distances = {point:np.inf for point in vertices}
distances[(0,0)] = 0
previous = {point:None for point in vertices}

# print((0, 0) in vertices)
# print(distances[(0,0)])
# print(graph[(0, 0)])

while vertices:
    minimum = np.inf
    u = None
    for vertex in vertices:
        if distances[vertex] < minimum:
            minimum = distances[vertex]
            u = vertex
    
    vertices = vertices - set([u])
       
    for adjacent in graph[u]:

        if adjacent[:2] not in vertices:
            continue
            
        alt = distances[u] + adjacent[2]
        if alt < distances[adjacent[:2]]:
            distances[adjacent[:2]] = alt
            previous[adjacent[:2]] = u
                        
# print(distances)
print(distances[(grid.shape[0] - 1, grid.shape[1] - 1)])
print(datetime.datetime.now() - start)
```
